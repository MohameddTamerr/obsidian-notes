## Step 1: Let Initial Population 

We Get It In First Time By Random

```python
import random

def generate_chromosome(length):
    chromosome = ""
    for i in range(length):
        chromosome += str(random.randint(0, 1))
    return chromosome

def generate_population(size, length):
    population = []
    for i in range(size):
        population.append(generate_chromosome(length))
    return population

population_size = 4
population = generate_population(population_size, chromosome_length)

print("Initial Population:")
for c in population:
    print(c)
```

## Step 2: Fitness Calculation

| Chromosome | Weight | Profit | Fitness |
|------------|--------|--------|---------|
| 101001     | 11     | 19     | 0       |
| 100011     | 9      | 20     | 20      |

```python
def fitness(chromosome):
    
    total_weight = 0
    total_profit = 0
    
    for i in range(len(chromosome)):
        
        if chromosome[i] == '1':
            total_weight += weights[i]
            total_profit += profits[i]
    
    if total_weight > max_weight:
        return 0
    
    return total_profit
```

## Step 4: calculate fitness

```python
fitness_values = []

for c in population:
    f = fitness(c)
    fitness_values.append(f)
    print(c, "Fitness =", f)
```

## Step 5: selection

```python
# combine chromosome and fitness
combined = list(zip(population, fitness_values))

# sort by fitness descending
combined.sort(key=lambda x: x[1], reverse=True)

# select best 2
parent1 = combined[0][0]
parent2 = combined[1][0]

print("Parent 1:", parent1)
print("Parent 2:", parent2)
```

## Step 7 Crossover By Single Crossover Point
Parent 1 = 101 | 001  
Parent 2 = 011 | 010  

Child 1 = 101010  
Child 2 = 011001

```python
def crossover(p1, p2):
    
    point = len(p1) // 2
    
    child1 = p1[:point] + p2[point:]
    child2 = p2[:point] + p1[point:]
    
    return child1, child2

child1, child2 = crossover(parent1, parent2)

print("Child 1:", child1)
print("Child 2:", child2)
```

## Step 7: mutation


Before mutation: 101010  

After mutation: 101110

```python
def mutation(chromosome):
    
    chromosome = list(chromosome)
    
    point = random.randint(0, len(chromosome)-1)
    
    if chromosome[point] == '0':
        chromosome[point] = '1'
    else:
        chromosome[point] = '0'
    
    return ''.join(chromosome)

child1 = mutation(child1)

print("Child 1 after mutation:", child1)
```

## Best Solution

Best chromosome = 100011  

Total weight = 9  
Total profit = 20

```python
print("Final Fitness:")

print(child1, fitness(child1))
print(child2, fitness(child2))
```
