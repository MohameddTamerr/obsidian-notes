## Task 1: PCA Implementation

```python
import numpy as np
import pandas as pd

# Step 1: Load dataset
df = pd.read_csv("data.csv")

# Remove non-numeric columns
df = df.drop(columns=["id", "diagnosis", "Unnamed: 32"])

# Convert dataframe to numpy array
X = df.values
```

```python
# Step 2: Standardize the data (mean centering)
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)
X_centered = (X - mean) / std
```

```python
# Step 3: Calculate covariance matrix
cov_matrix = np.cov(X_centered, rowvar=False)
```

```python
# Step 4: Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

```python
# Step 5: Sort eigenvalues and eigenvectors in descending order
sorted_index = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_index]
sorted_eigenvectors = eigenvectors[:, sorted_index]
```

```python
# Step 6: Select the top k eigenvectors (principal components)
k = 2
principal_components = sorted_eigenvectors[:, :k]
```

```python
# Step 7: Transform the dataset
X_pca = np.dot(X_centered, principal_components)

print("Manual PCA Result:")
print(X_pca)

print("\nTop Eigenvalues:")
print(sorted_eigenvalues[:k])
```

## Task 2: Using sklearn PCA

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Dataset
data = pd.read_csv("data.csv")

df = pd.DataFrame(data)
```

```python
# Drop non-numeric columns
df = df.drop(columns=["id", "diagnosis", "Unnamed: 32"])

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print("PCA Result:")
print(X_pca)

print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)
```
