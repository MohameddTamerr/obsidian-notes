## Solving ODEs with SciPy:

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the ODE
def model(y, t, k):
    dydt = -k * y
    return dydt

# Parameters
y0 = 100
k = 0.5
t = np.linspace(0, 10, 100)

# Solve ODE
y = odeint(model, y0, t, args=(k,))

# Plot
plt.plot(t, y, 'b-', label='Numerical Solution')
plt.plot(t, y0*np.exp(-k*t), 'r--', label='Analytical')

plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.show()
```
