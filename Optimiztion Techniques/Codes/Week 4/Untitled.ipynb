```python
import cvxpy as cp

# Decision variables
x1 = cp.Variable()
x2 = cp.Variable()

# Objective function
objective = cp.Maximize(40 * x1 + 30 * x2)

# Constraints
constraints = [
    2 * x1 + x2 <= 100,   # Machine hours
    x1 + 2 * x2 <= 80,    # Labor hours
    x1 >= 0,
    x2 >= 0
]

# Solve the problem
problem = cp.Problem(objective, constraints)
problem.solve()

# Results
print(f"Optimal x1 = {x1.value:.2f}")
print(f"Optimal x2 = {x2.value:.2f}")
print(f"Maximum profit = ${problem.value:.2f}")
```

```python
from scipy.optimize import minimize

# Objective: minimize x^2 + y^2
def objective(x):
    return x[0]**2 + x[1]**2

# Constraint: x + y = 10
def constraint(x):
    return x[0] + x[1] - 10

# Initial guess
x0 = [1, 1]

# Define constraint
con = {'type': 'eq', 'fun': constraint}

# Solve
solution = minimize(objective, x0, constraints=con)

# Print results
print(f"Optimal x = {solution.x[0]:.2f}")
print(f"Optimal y = {solution.x[1]:.2f}")
print(f"Minimum value = {solution.fun:.2f}")
```
