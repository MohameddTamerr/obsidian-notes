```python
import cvxpy as cp
```

```python
## -x+2*y<= 8,
## 2*x+y<= 14,
## 2*x-y<= 10,
## x>=0, x<=10,
## y>=0 , x<=10
```

```python
x=cp.Variable()
y=cp.Variable()
```

```python
constraints=[-x+2*y<= 8,2*x+y<= 14,2*x-y<= 10,x>=0, x<=10,y>=0 , x<=10]
```

```python
obj=cp.Maximize(x+y)
```

```python
prob=cp.Problem(obj,constraints)
```

```python
prob.solve()
```

```python
print("Status" ,prob.status)
print("Optimal Values" ,x.value,y.value)
print("Optimal Solution" ,prob.value)
```

```python
# max 40*x+30*y
# 2*x+y<=100,
# x+2*y<=80,
# y>=0

```

```python
x_new=cp.Variable()
y_new=cp.Variable()
```

```python
constraints_new=[2*x_new+y_new<=100,x_new+2*y_new<=80,y_new>=0,x_new>=0]
```

```python
obj_new=cp.Maximize(40*x+30*y)
prob_new=cp.Problem(obj_new,constraints_new)
```

```python
prob_new.solve()
```

```python
print("Status" ,prob_new.status)
print("Optimal Values" ,x_new.value,y_new.value)
print("Optimal Solution" ,prob_new.value)
```
