```python
import cvxpy as cp
```

```python
## -x+2*y<= 8,
## 2*x+y<= 14,
## 2*x-y<= 10,
## x>=0, x<=10,
## y>=0 , x<=10
```

```python
x=cp.Variable()
y=cp.Variable()
```

```python
constraints=[-x+2*y<= 8,2*x+y<= 14,2*x-y<= 10,x>=0, x<=10,y>=0 , x<=10]
```

```python
obj=cp.Maximize(x+y)
```

```python
prob=cp.Problem(obj,constraints)
```

```python
prob.solve()
```
